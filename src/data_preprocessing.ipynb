```python
import pandas as pd
```

### Read data

```python
weekly_testing_data = pd.read_excel("../data/raw_data/weekly_testing_data_EUEEAUK_2022-09-15.xlsx")
```

```python
weekly_testing_data.head()
```

### Extract relevant information

```python
# filter out data for germany
germany_data = weekly_testing_data[weekly_testing_data["country"] == "Germany"]
# filter out non-national data
germany_data = germany_data[germany_data["level"] == "national"]
# filter out data we don't need
germany_data = germany_data[["year_week", "tests_done", "positivity_rate"]]
# make 2 rows out of year and week
def get_year(year_week):
    return int(year_week.split("-")[0])
def get_week(year_week):
    return int(year_week.split("-")[1][1:])
germany_data["year"] = germany_data["year_week"].apply(get_year)
germany_data["week"] = germany_data["year_week"].apply(get_week)
germany_data.drop("year_week", axis=1,inplace=True)
# 
germany_data["time"] = germany_data["year"] + germany_data["week"]/53

germany_data.head()
```

### Remove bad values

```python
# remove NaN values
germany_data = germany_data.dropna()
germany_data.head()
```

### Save file

```python
# save data to csv
germany_data.to_csv("../data/preprocessed_data/germany_weekly_testing_data.csv", index=False)
```
